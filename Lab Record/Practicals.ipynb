```python
#Program to check if a year is a leap year or not

n = int(input("Enter a number: "))

if (n % 4 == 0 and n % 100 !=0) or (n % 400 == 0):
    print(f"{n} is a leap year.")
else:
    print(f"{n} is not a leap year.")    
```

```python
#Program to count the number of vowels 

text = input("Enter some text: ")
vowels = "aeiouAEIOU"
count = 0
for char in text:
    if char in vowels:
        count += 1

print(f"Number of vowels in the given text: {count}")
```

```python
#Program to reverse a number

num = int(input("Enter a number: "))
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

print("The reversed number is:", rev)
```

```python
#Program to find mean, median and mode of a given integers

n = int(input("How many integers? "))
numbers = []
for i in range(n):
    num = int(input(f"Enter number {i+1}: "))
    numbers.append(num)
print(numbers)

numbers.sort()
print("Sorted list is:", numbers)

mean = sum(numbers) / n
print("Mean is:", mean)

median = 0
if n % 2 == 0:
    median = (numbers[n//2 - 1] + numbers[n//2]) / 2
    print("Median :",median)
else:
    median = numbers[n//2]    
    print("Median :",median)

mode = max(set(numbers), key=numbers.count)
print("Mode is:", mode)    
```

```python
#Program to reverse vowels in a string

text = input("Enter a string: ")

vowels = "aeiouAEIOU"
text_list = list(text)  
vowel_chars = []

for ch in text_list:
    if ch in vowels:
        vowel_chars.append(ch)

vowel_chars.reverse()

j = 0
for i in range(len(text_list)):
    if text_list[i] in vowels:
        text_list[i] = vowel_chars[j]
        j += 1

reversed_text = "".join(text_list)

print("String after reversing vowels:", reversed_text)
```

```python
#Program to find the student with highest marks 

n = int(input("Enter number of students: "))

students = []

for i in range(n):
    name = input("Enter student name: ")
    marks = int(input("Enter marks: "))
    students.append((name, marks))  

topper = max(students, key=lambda x: x[1])


print("\nStudent with highest marks:")
print("Name :", topper[0])
print("Marks:", topper[1])
```

```python
#Program to read a CSV file containing employee data and calculate the average salary by department

import csv

filename = input("Enter CSV file name (with .csv extension): ")

departments = {}

with open(filename, 'r') as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        dept = row['department']
        salary = float(row['salary'])
       
        if dept not in departments:
            departments[dept] = []
        departments[dept].append(salary)

print("\nAverage Salary by Department:")

for dept, salaries in departments.items():
    avg_salary = sum(salaries) / len(salaries)
    print(f"{dept}: {avg_salary:.2f}")
```
