```python
import math
from shape.point import Point

class Circle:
    def __init__(self, center: Point, radius):
        self.center = center
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)

    def perimeter(self):
        return 2 * math.pi * self.radius

    def display(self):
        return f"Circle(Center: {self.center.display()}, Radius: {self.radius}), Area: {self.area()}, Perimeter: {self.perimeter()}" 
        
```
